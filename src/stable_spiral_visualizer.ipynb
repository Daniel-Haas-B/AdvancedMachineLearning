```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def dynamics(t, y, a, b, c, d):
    x, y = y
    dx_dt = a*x - b*y
    dy_dt = c*x + d*y
    return [dx_dt, dy_dt]

# Parameters
a = 1.0
b = 0.5
c = 0.5
d = 1.0

# Initial conditions
x0 = 1.0
y0 = 0.0
initial_conditions = [x0, y0]

# Time span
t_start = 0.0
t_end = 5000.0
t_eval = np.linspace(t_start, t_end, 100000)

# Numerical integration
solution = solve_ivp(dynamics, [t_start, t_end], initial_conditions, t_eval=t_eval, args=(a, b, c, d))

# Extracting solution
x_solution = solution.y[0]
y_solution = solution.y[1]

# Plotting trajectory
plt.plot(x_solution, y_solution)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trajectory of the System')
plt.grid(True)
plt.show()
```
