```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plot_utils
import seaborn as sns
```

```python
spiral_df = pd.read_csv('spiral.csv', header=0, names=['t', 'x', 'y', 'particle'])

#drop the particle column and time column
spiral_df = spiral_df.drop(['t'], axis=1)
```

```python
# 2d plot with particle as hue
fig, ax = plt.subplots(figsize=(10, 10))
# prevent it from filling the lines in between
df_particle_1 = spiral_df[spiral_df['particle'] == 1]
df_particle_2 = spiral_df[spiral_df['particle'] == 2]
df_particle_3 = spiral_df[spiral_df['particle'] == 3]
df_particle_4 = spiral_df[spiral_df['particle'] == 4]
df_particle_5 = spiral_df[spiral_df['particle'] == 5]
df_particle_6 = spiral_df[spiral_df['particle'] == 6]
df_particle_7 = spiral_df[spiral_df['particle'] == 7]
df_particle_8 = spiral_df[spiral_df['particle'] == 8]
df_particle_9 = spiral_df[spiral_df['particle'] == 9]


ax.plot(df_particle_1['x'], df_particle_1['y'], color='blue', linewidth=1.2, label='particle 1')
ax.plot(df_particle_2['x'], df_particle_2['y'], color='orange', linewidth=1.2, label='particle 2')
ax.plot(df_particle_3['x'], df_particle_3['y'], color='green', linewidth=1.2, label='particle 3')
ax.plot(df_particle_4['x'], df_particle_4['y'], color='red', linewidth=1.2, label='particle 4')
ax.plot(df_particle_5['x'], df_particle_5['y'], color='purple', linewidth=1.2, label='particle 5')
ax.plot(df_particle_6['x'], df_particle_6['y'], color='brown', linewidth=1.2, label='particle 6')
ax.plot(df_particle_7['x'], df_particle_7['y'], color='pink', linewidth=1.2, label='particle 7')
ax.plot(df_particle_8['x'], df_particle_8['y'], color='gray', linewidth=1.2, label='particle 8')
ax.plot(df_particle_9['x'], df_particle_9['y'], color='olive', linewidth=1.2, label='particle 9')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('2D Spiral')
ax.legend()
plt.show()

```
