```python
import sys
sys.path.append('../')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# import a code from outside the current directory

```

```python
SIGMA = 10 
BETA = 8/3
RHO = 28
```

```python
lorenz_df = pd.read_csv('../csvs/lorenz.csv', header=0, names=['t', 'x', 'y', 'z', 'particle'])
lorenz_df["px"] = -SIGMA * (lorenz_df["x"] - lorenz_df["y"])
lorenz_df["py"] = (RHO - lorenz_df["z"]) * lorenz_df["x"] - lorenz_df["y"]
lorenz_df["pz"] = lorenz_df["z"] * (lorenz_df["y"] - BETA)
lorenz_df["sumP"] = lorenz_df["px"]**2 + lorenz_df["py"]**2 + lorenz_df["pz"]**2 
```

```python
filename = 'lorenz_visualizer'

# Plot the Lorenz attractor second particle
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

traj_1 = lorenz_df[lorenz_df['particle'] == 0]
traj_2 = lorenz_df[lorenz_df['particle'] == 7]

traj_1_initial_condition = traj_1.iloc[0]
traj_2_initial_condition = traj_2.iloc[0]


ax.plot(traj_1['x'], traj_1['y'], traj_1['z'], label='IC 1' )
ax.plot(traj_2['x'], traj_2['y'], traj_2['z'], label='IC 2' )

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#ax.set_title('Lorenz Attractor')
ax.legend()
# remove the numbers from the axes
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

# Make x , y and z labels closer to the axes (default is 5.0)
ax.xaxis.labelpad = -10
ax.yaxis.labelpad = -10
ax.zaxis.labelpad = -10

plt.subplots_adjust(left=0, right=0.99, top=0.99, bottom=0) 


plt.show()
```
